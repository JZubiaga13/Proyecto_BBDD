### 1. TRATAMIENTO DE DATOS.

```python
'''Importar Bibliotecas.'''

import pandas as pd
```

```python
''' Cargar las tablas desde los csv'''

clase_1=pd.read_csv('./data/clase_1.csv',sep=';')
clase_2=pd.read_csv('./data/clase_2.csv',sep=';')
clase_3=pd.read_csv('./data/clase_3.csv',sep=';')
clase_4=pd.read_csv('./data/clase_4.csv',sep=';')
claustro=pd.read_csv('./data/claustro.csv',sep=';')
```

```python
''' Ver los diferentes datos.'''

display(clase_1,clase_2,clase_3,clase_4,claustro)
```

### 2. CREAR DATAFRAME.

```python
'''Crear DataFrame de Proyectos: Proyectos_DS y Proyectos_FS.'''

# Crear Dataframe Proyectos_DS
df_concatenado_ds = pd.concat(
    [clase_1[['Proyecto_HLF', 'Proyecto_EDA', 'Proyecto_BBDD', 'Proyecto_ML', 'Proyecto_Deployment' ]], 
     clase_2[['Proyecto_HLF', 'Proyecto_EDA', 'Proyecto_BBDD', 'Proyecto_ML', 'Proyecto_Deployment' ]]], 
    ignore_index=True
)

# Transformar las columnas en filas con melt
df_proyectos_ds = df_concatenado_ds.melt(var_name='Nombre')

# Asegurarse de que no existan espacios adicionales en la columna 'Nombre'
df_proyectos_ds['Nombre'] = df_proyectos_ds['Nombre'].str.strip()

#Elimina filas duplicadas basadas en la columna Nombre
df_proyectos_ds = df_proyectos_ds.drop_duplicates(subset=['Nombre'], keep='first').reset_index(drop=True)

# Crear la columna Vertical con el valor "DS"
df_proyectos_ds['Vertical'] = "DS"

# Reordenar las columnas para incluir la columna Vertical
df_proyectos1 = df_proyectos_ds[['Nombre', 'Vertical']]
df_proyectos1
```

```python
# Crear Dataframe Proyectos_FS
df_concatenado_fs = pd.concat(
    [clase_3[['Proyecto_WebDev', 'Proyecto_FrontEnd', 'Proyecto_Backend', 'Proyecto_React', 'Proyecto_FullStack']], 
     clase_4[['Proyecto_WebDev', 'Proyecto_FrontEnd', 'Proyecto_Backend', 'Proyecto_React', 'Proyecto_FullStack']]],
    ignore_index=True
)

# Transformar las columnas en filas con melt
df_proyectos_fs = df_concatenado_fs.melt(var_name='Nombre')

# Asegurarse de que no existan espacios adicionales en la columna 'Nombre'
df_proyectos_fs['Nombre'] = df_proyectos_fs['Nombre'].str.strip()

#Elimina filas duplicadas basadas en la columna Nombre
df_proyectos_fs = df_proyectos_fs.drop_duplicates(subset=['Nombre'], keep='first').reset_index(drop=True)

# Crear la columna Vertical con el valor "DS"
df_proyectos_fs['Vertical'] = "FS"

# Reordenar las columnas para incluir la columna Vertical
df_proyectos2 = df_proyectos_fs[['Nombre', 'Vertical']]
df_proyectos2
```

```python
# Crear Dataframe Proyectos, unificando los dos anteriores.

df_proyectos = pd.concat([df_proyectos1, df_proyectos2],ignore_index=True)

#Crear columna ID_proyecto.
df_proyectos['ID_proyecto']= range(1, len(df_proyectos) + 1)

# Reordenar las columnas para incluir la columna Vertical
df_proyectos = df_proyectos[['ID_proyecto', 'Nombre', 'Vertical']]

df_proyectos
```

```python
'''Creación del DataFrame de Alumnos.'''

# Crear cuatro datafreme por cada una de las clases.
df_alumnos1=pd.DataFrame(columns=['Nombre','Email','ID_campus','ID_curso'],index=range(0,len(clase_1)))
df_alumnos2=pd.DataFrame(columns=['Nombre','Email','ID_campus','ID_curso'],index=range(0,len(clase_2)))
df_alumnos3=pd.DataFrame(columns=['Nombre','Email','ID_campus','ID_curso'],index=range(0,len(clase_3)))
df_alumnos4=pd.DataFrame(columns=['Nombre','Email','ID_campus','ID_curso'],index=range(0,len(clase_4)))

# Realización de bucles para recorrer cada una de las filas e ir cargando los datos.
for i in range(len(clase_1)):
    df_alumnos1.iloc[i,0]=clase_1.loc[i,'Nombre']
    df_alumnos1.iloc[i,1]=clase_1.loc[i,'Email']
    df_alumnos1.iloc[i,2]=1
    df_alumnos1.iloc[i,3]=10

for i in range(len(clase_2)):
    df_alumnos2.iloc[i,0]=clase_2.loc[i,'Nombre']
    df_alumnos2.iloc[i,1]=clase_2.loc[i,'Email']
    df_alumnos2.iloc[i,2]=1
    df_alumnos2.iloc[i,3]=11

for i in range(len(clase_3)):
    df_alumnos3.iloc[i,0]=clase_3.loc[i,'Nombre']
    df_alumnos3.iloc[i,1]=clase_3.loc[i,'Email']
    df_alumnos3.iloc[i,2]=1
    df_alumnos3.iloc[i,3]=12

for i in range(len(clase_4)):
    df_alumnos4.iloc[i,0]=clase_4.loc[i,'Nombre']
    df_alumnos4.iloc[i,1]=clase_4.loc[i,'Email']
    df_alumnos4.iloc[i,2]=2
    df_alumnos4.iloc[i,3]=13

# Concatenar los diferentes dataframe creados anteriormente.
df_alumnos=pd.concat(
    [df_alumnos1,df_alumnos2,df_alumnos3,df_alumnos4],ignore_index=True
)

df_alumnos
```

```python
''' Crear DataFrame de Promoción.'''

df_promocion=pd.DataFrame(columns=['id_promocion','nombre','fecha_comienzo'])

# Crear columna llamada Fecha_comienzo.
df_promocion['fecha_comienzo']=[clase_1.loc[0,'Fecha_comienzo'],clase_2.loc[0,'Fecha_comienzo'],clase_3.loc[0,'Fecha_comienzo'],clase_4.loc[0,'Fecha_comienzo']]

# Crear columna llamada Nombre.
df_promocion['nombre']=['Septiembre','Febrero','Septiembre','Febrero']

# Crear columna llamada ID_promoción.
df_promocion['id_promocion']= range(1, len(df_promocion) + 1)

df_promocion
```

```python
'''Creación del DataFrame Campus.'''

# Concatenar las cuatro clases.
df_clasesp = pd.concat([clase_1,clase_2,clase_3,clase_4])

# Eliminar las columnas que no son necesarias.
df_clases = df_clasesp.drop(columns=['Fecha_comienzo','Proyecto_HLF', 'Proyecto_EDA', 'Proyecto_BBDD', 'Proyecto_ML',
       'Proyecto_Deployment', 'Proyecto_WebDev', 'Proyecto_FrontEnd',
       'Proyecto_Backend', 'Proyecto_React', 'Proyecto_FullStack'])

# Crear DataFrame mediante la columna de Campus usando los valores únicos.
df_campus = pd.DataFrame(df_clases['Campus'].unique())

# Renombrar la columna por el nombre ciudad.
df_campus = df_campus.rename(columns={0:'Ciudad'})

# Crear la columna ID_campus y crear su numeración.
df_campus['ID_campus']=range(1, len(df_campus) + 1)

# Reordenar las columnas.
df_campus=df_campus[['ID_campus','Ciudad']]

df_campus
```

```python
'''Creación del DataFrame Profesor.'''

# Recombrar el DataFrame de claustro con el nombre df_profesor.
df_profesor= claustro

# Crear columna ID_profesor y crear su numeración.
df_profesor['ID_profesor']=range(1, len(df_profesor) + 1)

# Elegir las columnas que nos interesan.
df_profesor=df_profesor[['ID_profesor','Nombre','Rol','Vertical', 'Campus']]   

# Renombrar columna Campus por Ciudad.
df_profesor=df_profesor.rename(columns={'Campus': 'ciudad'})
df_profesor=df_profesor.rename(columns={'ID_profesor':'id_profesor'})
df_profesor=df_profesor.rename(columns={'Nombre':'nombre'})
df_profesor=df_profesor.rename(columns={'Rol':'rol'})
df_profesor=df_profesor.rename(columns={'Vertical':'vertical'})

df_profesor
```

```python
'''Creación del DataFrame Aula.'''

# Crear DataFrame con unas columnas determinadas.
df_aula=pd.DataFrame(columns=['ID_aula','numero','ID_campus'])

# 2. Merge con la tabla auxiliar.
# Merge con alumnos para obtener ID_campus
df_aula= df_campus.merge(df_campus, on=["Ciudad"])

# Crear columna ID_aula y crear su numeración.
df_aula['ID_aula']=range(1, len(df_aula) + 1)

# Crear números de aulas, inventadas.
df_aula['numero']=[1,2]

# Eliminar columnas. 
df_aula=df_aula.drop(columns=['ID_campus_x', 'Ciudad'])

# Reordenar columnas.
df_aula= df_aula[['ID_aula', 'numero', 'ID_campus_y']]

# Cambiar nombre.
df_aula= df_aula.rename(columns={'ID_campus_y': 'ID_campus'})

df_aula
```

```python
'''Creación del DataFrame Profesor_curso.'''

# Crear DataFrame con unas columnas determinadas.
df_profesor_curso=pd.DataFrame(columns=['id_profesor','ID_curso'])

# Crear columna ID_aula y crear su numeración.
df_profesor_curso['ID_profesor_curso']=range(1, len(df_profesor_curso) + 1)

# Reordenar las columnas.
df_profesor_curso=df_profesor_curso[['ID_profesor_curso','id_profesor', 'ID_curso']]

# Borrar la columna ID_profesor_curso.
df_profesor_curso.drop(columns='ID_profesor_curso',axis=1,inplace=True)

# Crear columna nueva, forzando el tipo.
df_profesor_curso['ID_profesor']=df_profesor['id_profesor'].astype(int)

# Crear columna de ID_curso, con su numeración y su tipo.
df_profesor_curso['ID_curso']=[10,11,12,13,13,12,10,13,12,11]

df_profesor_curso['ID_curso']=df_profesor_curso['ID_curso'].astype(int)

df_profesor_curso
```

```python
'''Creación del DataFrame Curso.'''

df_curso=pd.DataFrame(columns=['ID_curso','vertical','modalidad','ID_campus','ID_aula','ID_promocion'])

# Crear columna modalidad y sus valores.
df_curso['modalidad']=['Presencial','Presencial','Online','Online']

# Crear columna ID_curso y crear su numeración.
df_curso['ID_curso']=range(1, len(df_curso) + 1)

df_curso
```

```python
#Cargar los datos de las columnas.
df_curso['vertical']=['DS','DS','FS','FS']
df_curso['ID_aula']=[1,2,2,1]
df_curso['ID_promocion']=[1,2,3,4]
df_curso['ID_campus']=[1,1,1,2]

df_curso
```

```python
'''Creación del DataFrame de Calificaciones.'''

# 1. Normalizar las tablas originales de las clases.
clase1_normalizada = clase_1.melt(
    id_vars=["Nombre", "Email", "Promoción", "Fecha_comienzo", "Campus"],
    var_name="Proyecto",
    value_name="resultado"
)

clase2_normalizada = clase_2.melt(
    id_vars=["Nombre", "Email", "Promoción", "Fecha_comienzo", "Campus"],
    var_name="Proyecto",
    value_name="resultado"
)

clase3_normalizada = clase_3.melt(
    id_vars=["Nombre", "Email", "Promoción", "Fecha_comienzo", "Campus"],
    var_name="Proyecto",
    value_name="resultado"
)

clase4_normalizada = clase_4.melt(
    id_vars=["Nombre", "Email", "Promoción", "Fecha_comienzo", "Campus"],
    var_name="Proyecto",
    value_name="resultado"
)

# Combinar ambas tablas normalizadas
tabla_normalizada = pd.concat([clase1_normalizada, clase2_normalizada, clase3_normalizada, clase4_normalizada])
tabla_normalizada
```

```python
# 2. Merge con las tablas auxiliares
# Merge con alumnos para obtener ID_alumno
df_calificaciones = tabla_normalizada.merge(df_alumnos, on=["Nombre", "Email"])

# Merge con proyectos para obtener ID_proyecto
df_calificaciones = df_calificaciones.merge(df_proyectos, left_on="Proyecto", right_on="Nombre")
df_calificaciones
```

```python
# 3. Generar la columna ID_calificacion
df_calificaciones["ID_calificacion"] = range(1, len(df_calificaciones) + 1)

# 4. Seleccionar columnas finales
df_calificaciones_final = df_calificaciones[["ID_calificacion", "resultado", "id_alumno", "Nombre_y"]]

# 5. Cambiar nombre columna Nombre_y por nombre_proyecto
df_calificaciones_final = df_calificaciones_final.rename(columns={"Nombre_y": "nombre_proyecto"})

df_calificaciones_final
```

### 3. CONVERTIR LOS DATAFRAME EN CSVs.

```python
'''Convertir los DataFrame en csv.'''

# Alumnos.
df_alumnos.to_csv('alumnos.csv', index=False)

# Proyectos.
df_proyectos.to_csv('proyectos.csv', index=False)

# Calificaciones.
df_calificaciones_final.to_csv('calificaciones.csv', index=False)

# Promoción.
df_promocion.to_csv('promocion.csv', index=False)

# Campus.
df_campus.to_csv('campus.csv', index=False)

# Aula.
df_aula.to_csv('aula.csv', index=False)

# Profesor.
df_profesor.to_csv('profesor.csv', index=False)

# Curso.
df_curso.to_csv('curso.csv', index=False)
```

```python
'''Bucle para rellenas las tablas de la BBDD.'''

# Recorrer las filas del DataFrame e insertar cada una de ellas en la tabla carrespondiente
for index, row in df_promocion.iterrows():
    insert_query = """
    INSERT INTO promocion (nombre, fecha_comienzo)
    VALUES (%s, %s);
    """
    # Pasar los valores de manera segura
    cursor.execute(insert_query, (row[1], row[2]))
# Confirmar la transacción
conn.commit()
```

```python
''' Carga de los datos en la tabla de Curso.'''

# Recorrer las filas del DataFrame e insertar cada una en la tabla.
for index, row in df_curso.iterrows():
    insert_query = """
    INSERT INTO curso (vertical, modalidad,id_campus,id_aula,id_promocion)
    VALUES (%s, %s,%s,%s,%s);
    """

    # Pasar los valores de manera segura
    cursor.execute(insert_query, (row[1], row[2],row[3], row[4],row[5]))
    
# Confirmar la transacción
conn.commit()
```

```python
''' Carga de los datos en la tabla de Alumnos.'''

# Recorrer las filas del DataFrame e insertar cada una en la tabla.
for index, row in df_alumnos.iterrows():
    insert_query = """
    INSERT INTO alumno (nombre_apellidos,email,id_campus,id_curso)
    VALUES (%s,%s,%s,%s);
    """

    # Pasar los valores de manera segura
    cursor.execute(insert_query, (row[0],row[1],row[2],row[3]))
    
# Confirmar la transacción
conn.commit()
```

```python
''' Carga de los datos en la tabla de Promoción.'''

# Recorrer las filas del DataFrame e insertar cada una en la tabla.
for index, row in df_promocion.iterrows():
    insert_query = """
    INSERT INTO promocion (nombre, fecha_comienzo)
    VALUES (%s, %s);
    """

    # Pasar los valores de manera segura
    cursor.execute(insert_query, (row[1], row[2]))
    
# Confirmar la transacción
conn.commit()
```

```python
''' Carga de los datos en la tabla de Profesor_curso.'''

# Recorrer las filas del DataFrame e insertar cada una en la tabla
for index, row in df_profesor_curso.iterrows():
    insert_query = """
    INSERT INTO profesor_curso (id_profesor,id_curso)
    VALUES (%s, %s);
    """

    # Pasar los valores de manera segura
    cursor.execute(insert_query, (int(row[0]), int(row[1])))

# Confirmar la transacción
conn.commit()
```
